```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data= pd.read_csv('/content/road accident data.csv', encoding='latin-1')
```

```python
df = pd.DataFrame(data)
df
```

```python
df['Accident Date'] = pd.to_datetime(df['Accident Date'])
accidents_over_time = df.groupby(df['Accident Date'].dt.date).size()
plt.figure(figsize=(12, 6))
accidents_over_time.plot()
plt.title('Number of Accidents Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Accidents')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='Casualty Severity', y='Age of Casualty')
plt.title('Distribution of Casualty Severity by Age')
plt.xlabel('Casualty Severity')
plt.ylabel('Age of Casualty')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='1st Road Class', hue='Road Surface')
plt.title('Road Surface Conditions by Road Class')
plt.xlabel('1st Road Class')
plt.ylabel('Number of Accidents')
plt.legend(title='Road Surface')
plt.grid(True)
plt.show()

```

```python
plt.figure(figsize=(12, 6))
sns.violinplot(data=df, x='Casualty Class', y='Age of Casualty', hue='Sex of Casualty', split=True)
plt.title('Distribution of Casualty Class by Age and Gender')
plt.xlabel('Casualty Class')
plt.ylabel('Age of Casualty')
plt.grid(True)
plt.show()
```

```python
# Select numeric columns
numeric_cols = ['Easting', 'Northing', 'Number of Vehicles', 'Age of Casualty']

# Compute correlation matrix
corr_matrix = df[numeric_cols].corr()

# Plot
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Heatmap of Numeric Variables')
plt.show()
```

```python
# Assume 'Number of Casualties' is calculated based on dataset's logic
# For simplicity, we use 'Number of Vehicles' as a proxy
sns.jointplot(data=df, x='Age of Casualty', y='Number of Vehicles', hue='Casualty Severity', palette='viridis')
plt.suptitle('Joint Plot of Age vs. Number of Casualties', y=1.02)
plt.show()
```

```python
# Plot pairplot
sns.pairplot(df, vars=numeric_cols, hue='Casualty Severity', palette='viridis')
plt.suptitle('Pairplot of Numeric Variables', y=1.02)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='Number of Vehicles', palette='viridis')
plt.title('Accidents by Number of Vehicles Involved')
plt.xlabel('Number of Vehicles')
plt.ylabel('Number of Accidents')
plt.grid(True)
plt.show()
```
